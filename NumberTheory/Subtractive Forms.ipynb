```python
def createSubMat(natNum):
    out = []
    for i in range(2,natNum):
        row = []
        res = natNum - i
        
        while res >= 0:
            row.append(res)
            res = res - i
            
        if(len(row) <= 2):
            break
        else:
            out.append(row)
        
    return out

def isPrime(natNum):
    for i in range(2,natNum-1):
        if(natNum%i == 0):
            return False
    return True
```

```python
def lastInEach(lsls):
    lscpy = lsls[:]
    return list(map(list.pop,lscpy))
```

```python
a = createSubMat(24)
lastInEach(a)
```

```python
for i in range(4,100):
    a = createSubMat(i)
    if(isPrime(i)):
        print(i,len(lastInEach(a)),"<--")
    else:
        print(i,len(lastInEach(a)))
```
