```python
import matplotlib.pyplot as plt
from ipywidgets import interact
```

```python
def isPrime(natNum):
    for i in range(2,natNum-1):
        if(natNum%i == 0):
            return False
    return True
```

```python
for i in range(2,100):
    val = [i % 2, i % 3, i % 4, i % 5, i % 6, i % 7, i % 8, i % 9]
    if(isPrime(i)):
        print(i,val,"<--")
    else:
        print(i,val)
```

```python
def vis1(a=1,b=1,c=1,d=1):
    out = []
    for i in range(2,100):
        val = sum([i % a, i%b, i%c,i%d])
        out.append(val)

    plt.plot(out)
    plt.show()

interact(vis1,a=(1,10),b=(1,10),c=(1,10),d=(1,10))
```

```python
def fitVal(fn,coefVec,rng = (1,101)):
    res = 0
    for i in range(rng[0],rng[1]):
        res += fn(i)-(i%coefVec[0]+i%coefVec[1]+i%coefVec[2])
    return res
```

```python
def fitXSquared(a,b,c):
    print(fitVal(lambda x: x**2,[a,b,c]))

interact(fitXSquared,a=(0,100),b=(0,100),c=(0,100))
```
